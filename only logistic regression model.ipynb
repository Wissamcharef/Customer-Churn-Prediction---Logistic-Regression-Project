```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["Churn"])
Y = df["Churn"]

x_train , x_test ,y_train , y_test = train_test_split(X , Y , test_size=0.2 ,random_state=42 )
print("x :" , x_train.shape)
print("y : ", y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import pandas as pd

# Logistic Regression model and hyperparameter grid
model_params = {
    'Logistic_Regression': {
        'model': LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced'),
        'params': {
            'C': [0.1, 1, 10],            # Inverse of regularization strength
            'solver': ['liblinear', 'lbfgs']
        }
    }
}

# Run Grid Search for Logistic Regression
scores = []

for model_name, mp in model_params.items():
    print(f"Training {model_name}...")
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, scoring='roc_auc', return_train_score=False)
    clf.fit(x_train, y_train)
    
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

# Create a DataFrame to display results
results_df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

print("\n--- Logistic Regression Results ---")
print(results_df)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, f1_score, accuracy_score
from sklearn.model_selection import learning_curve
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np

# Only Logistic Regression
model = LogisticRegression(C=1, solver='liblinear', class_weight='balanced', random_state=42)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
plt.subplots_adjust(wspace=0.3)

print("Processing Logistic Regression...")

# Train model
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_prob = model.predict_proba(x_test)[:, 1]

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False)
axes[0].set_title('Logistic Regression - Confusion Matrix')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

# Accuracy and F1 Score
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
axes[0].text(0.5, -0.2, f'Accuracy: {acc:.2f} | F1: {f1:.2f}', 
            ha='center', transform=axes[0].transAxes, fontweight='bold')

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

axes[1].plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
axes[1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
axes[1].set_xlim([0.0, 1.0])
axes[1].set_ylim([0.0, 1.05])
axes[1].set_xlabel('False Positive Rate')
axes[1].set_ylabel('True Positive Rate')
axes[1].set_title('Logistic Regression - ROC Curve')
axes[1].legend(loc="lower right")

# Learning Curve
train_sizes, train_scores, test_scores = learning_curve(
    model, x_train, y_train, cv=5, n_jobs=-1, 
    train_sizes=np.linspace(0.1, 1.0, 5), scoring='accuracy'
)

train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

axes[2].plot(train_sizes, train_mean, 'o-', color="r", label="Training score")
axes[2].plot(train_sizes, test_mean, 'o-', color="g", label="CV score")
axes[2].set_title('Logistic Regression - Learning Curve')
axes[2].set_xlabel('Training examples')
axes[2].set_ylabel('Accuracy')
axes[2].legend(loc="best")
axes[2].grid(True)

print("--- Logistic Regression Classification Report ---")
print(classification_report(y_test, y_pred))

plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Hyperparameter grid for Logistic Regression
param_grid = [
    {
        'solver': ['liblinear', 'saga'], 
        'penalty': ['l1'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'class_weight': [None, 'balanced']
    },
    {
        'solver': ['liblinear', 'newton-cg', 'lbfgs', 'saga'], 
        'penalty': ['l2'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'class_weight': [None, 'balanced']
    },
    {
        'solver': ['saga'],
        'penalty': ['elasticnet'],
        'C': [0.01, 0.1, 1, 10],
        'l1_ratio': [0.2, 0.5, 0.8], 
        'class_weight': [None, 'balanced']
    }
]

# Logistic Regression model
lr = LogisticRegression(max_iter=5000, random_state=42)

# Grid Search with multiple scoring metrics
grid_search = GridSearchCV(
    estimator=lr,
    param_grid=param_grid,
    cv=5,
    scoring={'f1': 'f1', 'precision': 'precision', 'accuracy': 'accuracy'},
    refit='f1',  
    n_jobs=-1,
    return_train_score=False
)

print("Starting extensive tuning...")
grid_search.fit(x_train, y_train)
print("WE FINISHED")

# Display best parameters and score
print(f"\nBest Parameters: {grid_search.best_params_}")
print(f"Best F1 Score: {grid_search.best_score_:.4f}")
```

```python
import pandas as pd

# Extract results from GridSearchCV
results = pd.DataFrame(grid_search.cv_results_)

# Select relevant columns
cols_to_keep = ['params', 'mean_test_f1', 'mean_test_precision', 'mean_test_accuracy', 'rank_test_f1']
results_df = results[cols_to_keep]

# Expand params dictionary into separate columns
params_df = pd.json_normalize(results_df['params'])
results_df = pd.concat([params_df, results_df.drop(columns=['params'])], axis=1)

# Rename columns for clarity
results_df = results_df.rename(columns={
    'mean_test_f1': 'F1 Score',
    'mean_test_precision': 'Precision',
    'mean_test_accuracy': 'Accuracy'
})

# Sort by F1 Score
results_df = results_df.sort_values(by='F1 Score', ascending=False).reset_index(drop=True)

# Display top 10 models
print("--- Top 10 Models ---")
print(results_df.head(10))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Get best Logistic Regression model
best_lr = grid_search.best_estimator_

print(f"\n--- Best Parameters Found ---")
print(grid_search.best_params_)

# Predictions on test set
y_pred = best_lr.predict(x_test)

# Classification report
print("\n--- Final Test Report ---")
print(classification_report(y_test, y_pred))

# Confusion matrix
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', cbar=False)
plt.title('Confusion Matrix (Best Logistic Regression Model)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Get predicted probabilities
y_prob = best_lr.predict_proba(x_test)[:, 1]

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

axes[0].plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
axes[0].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--') 
axes[0].set_xlim([0.0, 1.0])
axes[0].set_ylim([0.0, 1.05])
axes[0].set_xlabel('False Positive Rate')
axes[0].set_ylabel('True Positive Rate')
axes[0].set_title('ROC Curve (Best Logistic Regression Model)')
axes[0].legend(loc="lower right")
axes[0].grid(True, alpha=0.3)

# Learning Curve
train_sizes, train_scores, test_scores = learning_curve(
    best_lr, x_train, y_train, 
    cv=5, 
    n_jobs=-1, 
    train_sizes=np.linspace(0.1, 1.0, 10), 
    scoring='f1' 
)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

axes[1].plot(train_sizes, train_mean, 'o-', color="r", label="Training Score")
axes[1].plot(train_sizes, test_mean, 'o-', color="g", label="Cross-Validation Score")

axes[1].fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color="r")
axes[1].fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color="g")
axes[1].set_title('Learning Curve (Best Logistic Regression Model)')
axes[1].set_xlabel('Number of Training Examples')
axes[1].set_ylabel('F1 Score')
axes[1].legend(loc="best")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
